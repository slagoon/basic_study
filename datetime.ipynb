# dtaetime

```python
import datetime as dt
```

```python
# 현재 시각 출력
x = dt.datetime.now()
x
```

```python
# 요일 반환 (0:월, 1:화, 2:수, 3:목, 4:금, 5:토, 6:일)
x.weekday()
```

```python
x.strftime('%A %d. %B %Y')
```

```python
x.strftime('%H시 %M분 %S초')
```

```python
# 문자열을 datetime 클래스 객체로 전환
dt.datetime.strptime("2017-01-02 14:44", "%Y-%m-%d %H:%M")
```

```python
a= dt.datetime(1998,12,23)
a
```

```python
# strftime 메서드를 사용하여 자신의 생일을 “xxxx년 xx월 xx일 x요일”과 같은 형식으로 출력해보자.
a.strftime("%Y년 %m월 %d일")
```

## 날짜/시간 연산

```python
dt1 = datetime.datetime(2016, 2, 19, 14)
dt2 = datetime.datetime(2016, 1, 2, 13)
td = dt1- dt2
td
```

```python
td.days, td.seconds, td.microseconds
```

```python
td.total_seconds()
```

```python
t0 = datetime.datetime(2018, 9,1, 13)
d = datetime.timedelta(days=90, seconds = 3600)
```

```python
t0 + d
```

```python
# 특정일의 2달 후 날짜 구하기
from dateutil.relativedelta import relativedelta

t0 + relativedelta(months = 2)
```
